```python
# Define more synonyms with the help of NLTK WordNet
from nltk.corpus import wordnet

def generate_synonyms(uri, keyword, syns):
    for synonym in wordnet.synsets(keyword):
        for item in synonym.lemmas():
            if keyword != synonym.name() and len(synonym.lemma_names()) > 1:
                syns.append(item.name())
    return (uri, keyword, syns)
```

```python
for idx, (uri, keyword, syns) in enumerate(sympsyn):
    sympsyn[idx] = generate_synonyms(uri, keyword, syns)
```

```python
expanded_symp = []

def is_exist(keyword, collection):
    for idx,word,syns in collection:
        if keyword == word:
            return True

for idx,word,syns in sympsyn:
    for s in syns:
        if not is_exist(s, sympsyn):
            expanded_symp.append((uri, s))
    if not is_exist(word, sympsyn):
        expanded_symp.append((uri, word))
```

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
#@title Default title text
sentences = ["fever", "high body temperate", "lethargy"]

#Compute embedding for both lists
embedding_1= model.encode(sentences[0], convert_to_tensor=True)
embedding_2 = model.encode(sentences[1], convert_to_tensor=True)
embedding_3 = model.encode(sentences[2], convert_to_tensor=True)

# Calculate cosine similarity
print(util.pytorch_cos_sim(embedding_1, embedding_2))
print(util.pytorch_cos_sim(embedding_1, embedding_3))
print(util.pytorch_cos_sim(embedding_2, embedding_3))
```

```python
def get_similarity(ne, idx, word):
    embedding_1 = model.encode(ne, convert_to_tensor=True)
    embedding_2 = model.encode(word, convert_to_tensor=True)
    return (idx, word, util.pytorch_cos_sim(embedding_1, embedding_2))

nes = ["high body temperature"]

for idx, (uri, word) in enumerate(expanded_symp):
    expanded_symp[idx] = get_similarity(nes[0], uri, word)
```
